# 과거 당첨번호 수집 및 전처리

```python
import pandas as pd
import numpy as np

df = pd.read_excel("./data/로또당첨번호_1.xlsx", header = 1)
df
```

```python
# DB 칼럼 : 회차, 날짜, 순서, 당첨번호
use_col = ['회차', '추첨일', '당첨번호', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18']
df_his = df[use_col].iloc[1:]
df_his.columns = ['회차', '날짜', 1, 2, 3, 4, 5, 6, 7]
df_his.head()
```

```python
# 언피봇
df_db = pd.melt(df_his, id_vars=['회차', '날짜'], value_vars=[1, 2, 3, 4, 5, 6, 7],
        var_name= '순서', value_name='당첨번호')
df_db.head(20)
```

```python
df_db = df_db.sort_values(by=['날짜','순서'], ascending=[False,True])
df_db
```

```python
# 회차 소수점 없애기 + 나머지 변환
df_db = df_db.astype({
    '회차' : 'int',
    '날짜' : 'object',
    '순서' : 'int',
    '당첨번호': 'int'
})  
```

```python
df_db['회차'] = df_db['회차'].astype('int')
```

```python
import pymysql

conn = pymysql.connect(
         host = 'localhost'
        , port = 3306
        , user = 'root'
        , password = '1234'
        , database= '로또'
        , charset = 'utf8mb4'
        )
cursor = conn.cursor()
table_name = '당첨번호'
columns = ', '.join(df_db.columns)
placeholders = ', '.join(['%s'] * len(df_db.columns))
```

```python
# query
sql = f"insert into {table_name} ({columns}) values ({placeholders})"

# insert 
data = [tuple(row) for row in df_db.to_numpy()]
```

```python
# insert & 커밋 
cursor.executemany(sql, data)
conn.commit()
cursor.close()
conn.close()
```
