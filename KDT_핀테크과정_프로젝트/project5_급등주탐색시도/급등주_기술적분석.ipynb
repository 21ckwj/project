```python
df_go['ma5'] = df_go['Close'].rolling(5).mean()
df_go['ma20'] = df_go['Close'].rolling(20).mean()
df_go['ma60'] = df_go['Close'].rolling(60).mean()
df_go['ma120'] = df_go['Close'].rolling(120).mean()
```

```python
# 20일 표준편차
df_go['std20'] = df_go['Close'].rolling(20).std()
# 5일 표준편차
df_go['std5'] = df_go['Close'].rolling(5).std()
# 60일 표준편차
df_go['std60'] = df_go['Close'].rolling(60).std()
```

```python
# 볼린져 밴드
df_go['upper_b20'] = df_go['ma20'] + (df_go['std20']*2)
df_go['lower_b20'] = df_go['ma20'] - (df_go['std20']*2)
```

```python
# 5일볼린져 밴드
df_go['upper_b5'] = df_go['ma5'] + (df_go['std20']*2)
df_go['lower_b5'] = df_go['ma5'] - (df_go['std20']*2)

# 60일볼린져 밴드
df_go['upper_b60'] = df_go['ma60'] + (df_go['std60']*2)
df_go['lower_b60'] = df_go['ma60'] - (df_go['std60']*2)
```
