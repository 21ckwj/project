# ◆ AI (2005~2020)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
df_ai = pd.read_csv('./data/AI타겟추가.csv',index_col=0)
df_ai = df_ai.set_index('시간')
df_ai
```

```python
# 2019년은 test를 위해 빼놓음

df_ai1 = df_ai.loc['2005-01-01':'2018-12-31']
df_ai1
```

## (1) 일주일 선행

```python
y = df_ai1['AI발생여부'][:-7]
x = df_ai1.drop(columns = ['AI발생여부']).shift(7)['2005-01-08':]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=111)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

from sklearn.metrics import classification_report, f1_score, roc_auc_score

svc = SVC(probability=True)
rfc = RandomForestClassifier()
ada = AdaBoostClassifier()
gbm = GradientBoostingClassifier()

xgb = XGBClassifier()
lgbm = LGBMClassifier()
cat = CatBoostClassifier()

model = [svc, rfc, ada, gbm,xgb, cat]
```

```python
model = [svc, rfc, ada, gbm,xgb, cat]
for m in model:
    m.fit(x_train,y_train)
    preds = m.predict(x_test)
    
    preds = m.predict(x_test)
    proba = m.predict_proba(x_test)
    
    train_preds = m.predict(x_train)    
    test_preds = m.predict(x_test)

    train_f1 = f1_score(y_train, train_preds)
    test_f1 = f1_score(y_test, test_preds)
    
    train_score = m.score(x_train, y_train)
    test_score = m.score(x_test, y_test)
    
    print(m.__class__.__name__)
    print(classification_report(y_test, test_preds))
    
    print('train_f1 : ', train_f1)
    print('test_f1 : ', test_f1)
    print('train_score : ', train_score)
    print('test_score : ', test_score)
 
    print(roc_auc_score(y_test,proba[:,1]))
    print('-'* 50)
```

## scaler

```python
# scaler
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

for m in model:
    m.fit(x_train_scaled,y_train)
    preds = m.predict(x_test_scaled)
    
    print(m.__class__.__name__)
    print(classification_report(y_test, preds))
    print(roc_auc_score(y_test,m.predict_proba(x_test_scaled)[:,1]))
    print('-'* 50)
```

## SMOTE

```python
# smote 사용
import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=111)
x_train_sm, y_train_sm = smote.fit_resample(x_train_scaled, y_train)
for m in model:
    m.fit(x_train_sm,y_train_sm)
    preds = m.predict(x_test_scaled)
    proba = m.predict_proba(x_test_scaled)
    
    print(m.__class__.__name__)
    print(classification_report(y_test, preds))
    print(roc_auc_score(y_test,proba[:,1]))
    print('-'* 50)
```
