## 타겟 기준 수정: 국내 전체

```python
df_all = pd.read_csv('./data/산불타겟_국내.csv')
df_all
```

```python
# 발생시 1 넣어줄 [1,1,1,...1] 
answer = np.repeat(1, len(df_all))

# 데이터프레임만들기
target_lst = df_all['발생일시']
df_target = pd.DataFrame({'시간':target_lst, '산불발생여부':answer})
df_target.value_counts()
```

```python
# 날씨 인덱스에서 원위치 후 합치기
df_c2.reset_index(inplace=True)
df_model = pd.merge(df_c2, df_target, how='left' ,on='시간')

# 타겟값 1 나머지 0 
df_model['산불발생여부'] = df_model['산불발생여부'].apply(lambda x : x if x == 1 else 0)
# df_model.drop(columns = ['index'], inplace=True)
df_model
```

```python
df_model.to_csv('./data/산불타겟추가_피해면적기준.csv')
```

## 타겟 기준 수정: 발생 가장 많은 지역

```python
# 발생시 1 넣어줄 [1,1,1,...1] 
answer = np.repeat(1, len(df_kb))

# 데이터프레임만들기
target_lst = df_kb['발생일시']
df_target = pd.DataFrame({'시간':target_lst, '산불발생여부':answer})
df_target.value_counts()
```

```python
# 날씨 인덱스에서 원위치 후 합치기
df_c2.reset_index(inplace=True)
df_model = pd.merge(df_c2, df_target, how='left' ,on='시간')

# 타겟값 1 나머지 0 
df_model['산불발생여부'] = df_model['산불발생여부'].apply(lambda x : x if x == 1 else 0)
df_model.drop(columns = ['index'],inplace=True)
df_model
```
