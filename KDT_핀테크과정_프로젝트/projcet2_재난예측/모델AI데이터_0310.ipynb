```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/조류인플루엔자 발생현황.csv',index_col=0)
df
```

```python
df.columns = ['가축전염병명','농장명(농장주)','농장소재지','발생일자','축종(품종)','발생두수(마리)','진단기관','종식일']
```

```python
df.head()
```

```python
# 날짜 처리
```

```python
# 결측값 확인
df['발생일자'].isnull().sum()
```

```python
df['발생일시']= df['발생일자'].apply(lambda x : x.split(' ')[0])
df['발생일시']
```

```python
# 지역처리
```

```python
df['농장소재지'].unique()
```

```python
df['발생장소'] = df['농장소재지'].apply(lambda x : x.split(' ')[0]+' '+ x.split(' ')[1] )
df['발생장소']
```

```python
df.head()
```

```python
df.to_csv('./data/조류인플루엔자_1.csv',index=False)
```

# ◆ 조류인플루엔자(AI) 타깃 정하기

```python
df_ai = pd.read_csv('./data/조류인플루엔자_1.csv')
df_ai
```

```python
df_ai['발생장소_광역시'] = df_ai['발생장소'].apply(lambda x : x.split(' ')[0])
df_ai.head()
```

```python
df_ai['발생장소_광역시'].value_counts() # 지역별로 조사하기엔 데이터 수가 너무 작음
```

```python
# 발생시 1 넣어줄 [1,1,1,...1] 
answer = np.repeat(1, len(df_ai))

# 데이터프레임만들기
target_lst = df_ai['발생일시']
df_target = pd.DataFrame({'시간':target_lst, 'AI발생여부':answer})
df_target.value_counts()
```

```python
# 중복날짜 제거
df_target.drop_duplicates(inplace=True)
df_target.value_counts()
```

```python
# 구간(2005년 ~2021년8월)
df_c2 = df_c1.loc['2005-01-01':'2021-08-31']
df_c2 
```

## 기후 데이터 + AI타겟 합치기

```python
# 합치기
df_model = pd.merge(df_c2, df_target, how='left' ,on='시간')

# 타겟값 1 나머지 0 
df_model['AI발생여부'] = df_model['AI발생여부'].apply(lambda x : x if x == 1 else 0)
df_model
```

```python
df_model.to_csv('./data/AI타겟추가.csv')
```
