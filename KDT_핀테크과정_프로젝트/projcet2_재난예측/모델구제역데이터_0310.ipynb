# 국내,고층,주변국 데이터(df_c) + 자외선(df_u)

```python
import pandas as pd
import numpy as np
```

```python
df_c = pd.read_csv('./data/국내+고층+주변국.csv')
df_c # 1996년 ~ 21년 8월
```

```python
# 자외선 데이터

df_u = pd.read_csv('./data/자외선최종(2005-2020).csv')
df_u.tail()
```

```python
# 보간법으로 채우기

df_u['자외선 B 일최고값'] = df_u['자외선 B 일최고값'].interpolate()
df_u.isnull().sum()
```

```python
# 자외선 추가하기

df_c1 = pd.merge(df_c,df_u, how='left', on='시간' )
df_c1
```

### df_c : 국내+고층+주변국
### df_c1: 국내+고층+주변국+자외선

==================================================================================

## df_c2 : 통합기후+자외선(df_c1) 기간설정(2005 ~ 2019)

```python
# 인덱스 설정
df_c1 = df_c1.set_index('시간')

# 구간(2000년 ~2019년)
df_c2 = df_c1.loc['2005-01-01':'2019-12-31']
df_c2
```

```python
df_c2.isnull().sum().sum()
```

# 구제역 전처리

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/구제역 발생현황.csv',index_col=0)
df
```

```python
df.columns = ['가축전염병명','농장명(농장주)','농장소재지','발생일자','축종(품종)','발생두수(마리)','진단기관','종식일']
```

```python
df.head()
```

```python
# 날짜 처리
```

```python
# 결측값 확인
df['발생일자'].isnull().sum()
```

```python
df['발생일시']= df['발생일자'].apply(lambda x : x.split(' ')[0])
df['발생일시']
```

```python
# 지역처리
```

```python
df['농장소재지'].unique()
```

```python
df['발생장소'] = df['농장소재지'].apply(lambda x : x.split(' ')[0]+' '+ x.split(' ')[1] )
df['발생장소']
```

```python
df.head()
```

```python
df.to_csv('./data/구제역_1.csv', index=False)
```

## 타겟 뽑기(전국)

```python
df_gu = pd.read_csv('./data/구제역_1.csv')
df_gu
```

```python
df_gu['발생일시'].values
```

```python
# 발생시 1 넣어줄 [1,1,1,...1] 
answer = np.repeat(1, len(df_gu))

# 데이터프레임만들기

target_lst = df_gu['발생일시']
df_target = pd.DataFrame({'시간':target_lst, '구제역발생여부':answer})
df_target.value_counts()
```

```python
# 중복날짜 제거
df_target.drop_duplicates(inplace=True)
df_target.value_counts()
```

# 기후데이터 + 구제역 타겟 합치기

```python
# 합치기
df_model = pd.merge(df_c2, df_target, how='left' ,on='시간')

# 타겟값 1 나머지 0 
df_model['구제역발생여부'] = df_model['구제역발생여부'].apply(lambda x : x if x == 1 else 0)
df_model
```

```python
df_model.to_csv('./data/구제역타겟추가.csv')
```
