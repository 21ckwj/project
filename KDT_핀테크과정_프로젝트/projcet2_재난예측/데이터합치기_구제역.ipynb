```python
import pandas as pd
import numpy as np
```

```python
# 구제역 데이터
df_gu = pd.read_csv('./data/구제역_1.csv')
df_gu.head()
```

```python
# merge를 위해 추가
df_gu['시간'] = df_gu['발생일시']
# 지역별 기후데이터를 위해 추가
df_gu['지역'] = df_gu['농장소재지'].apply(lambda x: x.split(' ')[0])
# feature값 부여
df_gu['구제역발생'] = np.repeat(1, len(df_gu))
df_gu.head(2)
```
