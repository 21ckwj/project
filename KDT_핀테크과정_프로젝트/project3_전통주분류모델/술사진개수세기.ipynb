```python
import pandas as pd
import numpy as np
import os
```

```python
# drinks 폴더안에 파일 개수
path = './drinks/'
drink_lst = os.listdir(path)
len(drink_lst)
```

```python
# 술 이미지 개수
lst = []
name= []
for d in drink_lst:
    path1 = path + d
    drink_image = os.listdir(path1)
    name.append(d)
    lst.append(len(drink_image))
    
    print('술이름: ',d)
    print('이미지갯수: ',len(drink_image))
    print('-'* 50)
df = pd.DataFrame({'술이름':name, '이미지갯수': lst})
df
```
