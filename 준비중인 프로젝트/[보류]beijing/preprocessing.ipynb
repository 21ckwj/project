# 주기 분석

## 1. fourier transformation

```python
# Nan값 채우기
series = df_pm['pm2.5'].fillna(method='ffill').dropna()
sp =np.fft.fft(series)
freq = np.fft.fftfreq(len(series))

fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
ax.plot(freq, sp.real, lw=1, c="purple")
# ax.set_xlim(-0.003, 0.2)
ax.set_xlim(-0.005, 0.1)
# ax.set_ylim(-7e6, 1e7)
ax.set_ylim(-7e6, 0.5 * 1e7)
ax.set_title("Fourier Transform")
```

```python
series = df_pm['pm2.5'].fillna(method='ffill').dropna()
sp =np.fft.fft(series)
freq = np.fft.fftfreq(len(series))

fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
ax.plot(freq, sp.real, lw=1, c="purple")
# ax.set_xlim(-0.003, 0.2)
ax.set_xlim(-0.005, 0.05)
# ax.set_ylim(-7e6, 1e7)
ax.set_ylim(-7e6, 0.5 * 1e7)
ax.set_title("Fourier Transform")

# 연단위, 6개월 단위, ..
ax.axvline(1/(365*24), c="orange", lw=5, alpha=0.5, zorder=0)
ax.axvline(1/365, c="orange", lw=5, alpha=0.5, zorder=0)
ax.axvline(1/30, c="orange", lw=5, alpha=0.5, zorder=0)
# ax.axvline(1/7, c="orange", lw=5, alpha=0.5, zorder=0)

ax.text(1/(365*24),0.3*1e7, r"$8760^{-1}$", ha="center", color="k", fontweight="bold")
ax.text(1/365, 0.1*1e7, r"$365^{-1}$", ha="center", color="k", fontweight="bold")
ax.text(1/30, 0.2*1e7, r"$30^{-1}$", ha="center", color="k", fontweight="bold")
# ax.text(1/7, 0.3*1e7, r"$\d7^{-1}$", ha="center", color="k", fontweight="bold")
```
