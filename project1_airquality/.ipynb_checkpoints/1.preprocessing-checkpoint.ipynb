```python
# 빈 칼럼 삭제
df = data.copy()
df = df.drop(columns=['Unnamed: 15','Unnamed: 16'])
df
```

```python
# Nan값 제거
df = df.dropna()
df
```

```python
# 날짜 시간 합친 인덱스 설정
df['Time'] = df['Time'].str.replace(".",":")
df['date'] = pd.to_datetime(df['Date'] +" "+ df['Time'])
df = df.drop(columns=['Date','Time'])
df = df.set_index('date')
df.head()
```
